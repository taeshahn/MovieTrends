```
import requests
from bs4 import BeautifulSoup
import pandas as pd
import datetime

#URL 정상 여부 확인 func
def get_html(url, next_day) :
    html = ""
    resp = requests.post(url,data={'loadEnd': '0' , 'searchType': 'search', 'sSearchFrom': next_day,'sSearchTo': next_day})

    if resp.status_code == 200 :
        html = resp.text

    return html

day = datetime.datetime(2019,1,1)           # 시작일자 지정 (2019년 1월 1일)
data = []                                   # 데이터를 저장할 리스트

for dd in range(0, 2):         # range(0 , x) 요기서 x 값이 조회 기간을 의미 지금은 2로 설정되어 이틀만 조회
    next_day = (day + datetime.timedelta(days=dd)).strftime('%Y-%m-%d')     # 일자 계산 및 문자 재변환

    # 우리가 찾고자하는 주소와 함께 변환한 날짜를 post로 param 던짐
    c = get_html('http://www.kobis.or.kr/kobis/business/stat/boxs/findDailyBoxOfficeList.do',next_day)
    soup = BeautifulSoup(c)

    s_table = soup.find_all("tr")
    s_table




    index = []                                  # 데이터가 뒤죽박죽이라 td 전용 리스트

    for idx, tr in enumerate(s_table):              # tr 추출을 위한 loop
            if idx > 0:
                tds = list(tr.find_all('td'))           # td  전부 찾아서 넣어주고
                for ix, td in enumerate(tds):           # td 추출을 위한 loop
                    text = tds[ix].text.strip()           # 항목 하나씩 문자로 치환
                    index.append(tds[ix].text.strip())    # 문자를 리스트 1개로 append 시키기
                index.insert(0, next_day)               # 조회(기준)일자를 넣음
                data.append(index)                      # 1개의 row 값 리스트를 데이터로 넣어주고
                index = []                              # 1개의 row 값을 reset

data
```
