```python
import time
import pandas as pd
import os
import requests
import urllib
from bs4 import BeautifulSoup
import re
```

```python
cnt = 0
page = 0
job_seq = 0
start_dt = '2019-03-23'
end_dt = '2019-06-23'
orgin_movie_nm = '어벤져스 엔드게임'
movie_nm = urllib.parse.quote(orgin_movie_nm, encoding='CP949')
```

```python
find_url = 'https://cafe.naver.com/ArticleSearchList.nhn?search.clubid=10000052&search.searchdate='+str(start_dt)+str(end_dt)+'&search.searchBy=1&search.query='+movie_nm+'&search.defaultValue=1&search.includeAll=&search.exclude=&search.include=&search.exact=&search.sortBy=date&userDisplay=50&search.media=0&search.option=0'

resp = requests.get(find_url)
content = resp.content.decode('ms949')
iframe = BeautifulSoup(content, "html.parser").find("div", {"class":"skin-1080 fl"})
trs = iframe.find_all('tr', {'class': None, 'id': None})

m = [i   for i in range(0, len(trs), 2)]
n = [i+1 for i in range(0, len(trs), 2)] # 불필요

rslt = pd.DataFrame()
for m, n in zip(m, n):
    s = trs[m].find('td', {'class': 'td_article'}).find('a', {'class':'article'})
    t = trs[m].find('a', {'class': 'm-tcol-c'})
    u = trs[m].find('em', {'class': None})
    v = trs[m].find('td', {'class': 'td_date'})
    w = trs[m].find('td', {'class': 'td_view'})

    title = re.sub('\n|&lt|&gt|;|\s{2,}', '', re.sub('<[^>]*>', '', str(s)))
    author = re.sub('\n|&lt|&gt|;|\s{2,}', '', re.sub('<[^>]*>', '', str(t)))
    comments = re.sub('\n|&lt|&gt|;|\s{2,}', '', re.sub('<[^>]*>', '', str(u)))
    date = re.sub('\n|&lt|&gt|;|\s{2,}', '', re.sub('<[^>]*>', '', str(v)))
    view = re.sub('\n|&lt|&gt|;|\s{2,}', '', re.sub('<[^>]*>', '', str(w)))
    
    print("\n====================================\n" + title + "\n" + author + "\n" + comments + "\n" + date + "\n" + view)
    
    new_row = {'title': title, 'author': author, 'comments': comments, 'date': date, 'view': view}
    rslt = rslt.append(new_row, ignore_index = True)
```
