# Google Trends Crawling

```python
from pytrends.request import TrendReq
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pytrends = TrendReq(hl='ko-KR', tz=360)

keyword = ['Avengers', 'Beauty and the Beast', "Parasite"]
pytrend = TrendReq()
pytrend.build_payload(kw_list=keyword, timeframe='today 5-y')
# Interest Over Time
trends = pytrend.interest_over_time().iloc[:,0:len(keyword)]
trends.reset_index(level=0, inplace=True)
trends = pd.melt(trends, id_vars="date")
```

```python
trends
```

```python
mpl.rc('font', family='NanumGothic')

ax = sns.lineplot(x='date', 
                  y='value', 
                  hue='variable',
#                   size='Size',
                  data=trends)

plt.title('Line Graph by seaborn', fontsize=20)
plt.ylabel('Trend', fontsize=14)
plt.xlabel('Date', fontsize=14)
plt.legend(fontsize=12, loc='best')
plt.show()
```
